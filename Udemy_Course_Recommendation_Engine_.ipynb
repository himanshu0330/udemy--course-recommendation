```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Cosine Similairty Matrix
# Formula = Similarity = (A.B) / (||A||.||B||)
similarity = cosine_similarity(vectors)
```

# Making a General Function to recommend course

```python
def recommend_course(title,records=5):
 idx = course_indices[title]
 scores = list(enumerate(similarity[idx]))
 sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)
 selected_course_indices = [i[0] for i in sorted_scores[1:]]
 selected_course_scores = [i[1] for i in sorted_scores[1:]]
 result = df['course_title'].iloc[selected_course_indices]
 rec_df = pd.DataFrame(result)
 rec_df['similarity_scores'] = selected_course_scores
 return rec_df.head(records)
```

# Task :- Similairty between User Input and Course title

```python
def calculate_confidence(course_title, user_input):
    course_title_lower = course_title.lower()
    user_input_lower = user_input.lower()
    keywords = course_title_lower.split()
    num_keywords = len(keywords)
    # Count how many keywords are present in the user input
    matches = sum(keyword in user_input_lower for keyword in keywords)
    # Calculate confidence as the percentage of keywords present in the user input
    confidence = (matches / num_keywords) * 100
    return confidence
# Example usage:
course_title = "Introduction to Python Programming"


user_input = input("Enter your input: ")
for i in df['course_title']:
  confidence = calculate_confidence(i, user_input)
  if confidence > 50:
    print(f"Confidence level: {confidence:.2f}%, {i}")
print(confidence)
```
